### Task 1

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import r2_score, roc_auc_score, roc_curve, auc
```

```python
boston = load_boston()

data = boston["data"]

data
```

```python
feature_names = boston["feature_names"]

feature_names
```

```python
target = boston["target"]

target
```

```python
X = pd.DataFrame(data, columns=feature_names)
X
```

```python
y = pd.DataFrame(target, columns=["price"])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)

y_lr_pred = lr.predict(X_test)

R2_lf = r2_score(y_test, y_lr_pred)
R2_lf
```

### Task 2

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)

model.fit(X_train, y_train.values[:, 0])

y_rf_pred = model.predict(X_test)

R2_rf = r2_score(y_test, y_rf_pred)
R2_rf
```

```python
'Модель ' + ('"линейная регресия"' if R2_lf > R2_rf else '"случайный лес"') + ' лучше'
```

### Task 4

```python
df = pd.read_csv('creditcard.csv', delimiter=',')
df
```

```python
s = df.value_counts(normalize=True)
s.shape
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
df.head(10)
```

```python
X = df.drop('Class', axis=1)
X
```

```python
y = df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15], 'max_features': np.arange(3, 5), 'max_depth': np.arange(4, 7)}]
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
y_pred_proba = model.predict_proba(X_test)[:,1]
y_pred_proba
```

```python
roc_auc_score(y_test, y_pred_proba)
```
