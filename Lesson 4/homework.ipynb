### Task 2

```python
t = np.linspace(0, 10, 51)

t
```

```python
f = np.cos(t)

f
```

```python
plt.plot(t, f, 'g')

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.title('График f(t)')

plt.xlabel('Значения t')
plt.ylabel('Значения f')

plt.show()
```

### Task 3

```python
x = np.linspace(-3, 3, 51)

x
```

```python
y1 = x ** 2

y1
```

```python
y2 = 2 * x + 0.5

y2
```

```python
y3 = -3 * x - 1.5

y3
```

```python
y4 = np.sin(x)

y4
```

```python
fig = plt.figure(figsize=(8, 6))

gs = fig.add_gridspec(2, 2, hspace=0.3, wspace=0.3)

ax = gs.subplots()

ax[0, 0].plot(x, y1)
ax[0, 0].axis(xmin=-5, xmax=5)
ax[0, 0].set_title('График y1')

ax[0, 1].plot(x, y2)
ax[0, 1].set_title('График y2')

ax[1, 0].plot(x, y3)
ax[1, 0].set_title('График y3')

ax[1, 1].plot(x, y4)
ax[1, 1].set_title('График y4')

plt.show()
```

### Task 4

```python
df = pd.read_csv('creditcard.csv', delimiter=',')

df
```

```python
counts = df['Class'].value_counts(dropna=False)

counts
```

```python
counts.plot.bar()

plt.show()
```

```python
counts_log = np.log(counts)

counts_log
```

```python
counts_log.plot.bar()

plt.show()
```

```python
plt.hist(df[df['Class'] == 1]['V1'], bins = 20, alpha = 0.5, label = 'Class 1', color = 'red', density = True)
plt.hist(df[df['Class'] == 0]['V1'], bins = 20, alpha = 0.5, label = 'Class 0', color = 'gray', density = True)

plt.legend(loc='upper left')

plt.xlabel('V1')

plt.show()
```
